```python
import pandas as pd
import numpy as np
import hycharts as hc
import hycharts.threedee as hc3
```

```python
as_df = pd.DataFrame(np.random.randn(1000,3),columns=['a','b','c'])
```

```python
chart = hc.hist(as_df['a'])
chart.draw()
```

```python
chart = hc.line(as_df['a'][:30])
chart.draw()
```

```python
chart = hc.line(as_df[:30])
chart.draw()
```

```python
chart = hc.scatter(as_df['a'],as_df['b'])
chart.draw()
```

```python
as_df = pd.DataFrame(np.random.randint(low=1000,high=1000000,size=(10,4)), columns=['a','b','c','d'], index=np.arange(1990,2000))
```

```python
chart = hc.area(as_df)
chart.draw()
```

```python
as_df = pd.DataFrame(np.random.randn(250,3),columns=['a','b','c'])
chart = hc3.scatter(as_df['a'],as_df['b'],as_df['c'], colorbypoint=True)
chart.draw()
```
