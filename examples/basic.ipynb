```python
import pandas as pd
import numpy as np
from hycharts import HighChart
```

```python
as_df = pd.DataFrame(np.random.randn(1000,3),columns=['a','b','c'])
```

```python
chart = HighChart.hist(as_df['a'])
chart.draw()
```

```python
chart = HighChart.line(as_df['a'][:30])
chart.draw()
```

```python
chart = HighChart.scatter(as_df['a'],as_df['b'])
chart.draw()
```
